## 1. Carregando os Cenarios

```python
grafo_a, bst_a = carregar_grafo_json('../data/raw/cenario_a_rs.json')
grafo_b, bst_b = carregar_grafo_json('../data/raw/cenario_b_matopiba.json')
print('Cenario A:', grafo_a)
print('Cenario B:', grafo_b)
```

## 2. BST — Municipios por Risco

```python
print('--- Cenario A: ordem crescente de risco ---')
for v in bst_a.percurso_in_order():
    print(f'  {v[IDX_NOME]:<22} risco={v[IDX_RISCO]:.2f}')

print('\n--- Municipios com risco >= 0.70 ---')
for v in bst_a.buscar(0.70, 1.0):
    print(f'  {v[IDX_NOME]:<22} risco={v[IDX_RISCO]:.2f}')
```

## 3. Dijkstra — Plano de Atendimento

```python
dcp_a = DijkstraComPrioridade(grafo_a, bst_a)
dcp_a.imprimir_plano(origem=4314902, limiar_risco=0.70)

print()
dcp_b = DijkstraComPrioridade(grafo_b, bst_b)
dcp_b.imprimir_plano(origem=1721000, limiar_risco=0.75)
```

## 4. Forca Bruta vs Dijkstra

```python
fb = ForcaBruta(grafo_a)
cam, custo_fb = fb.buscar_todos_caminhos(4314902, 4306403)
print(fb.relatorio(4314902, 4306403))

dij = Dijkstra(grafo_a)
cam_d, custo_d = dij.caminho_minimo(4314902, 4306403)
print(f'\nDijkstra: {custo_d:.2f} h | Gap: {abs(custo_d-custo_fb)/custo_fb*100:.4f}%')
```
